```python
#creation of set
set1 = {"aanjneya","satvik","aarshabh"}
print(set1)
```

```python
#duplicates not allowed
set1 = {"aanjneya","satvik","aarshabh","aanjneya"}
print(set1)
```

```python
#the values true and 1 are concidered same
set1 = {"aanjneya","satvik","aarshabh",True,1,2}
print(set1)
```

```python
#set with strings integers and boolean values
set1={"aanjneya",56,True,78,"kai",123.54}
set1
```

```python
#searching option
set1 = {"aanjneya","satvik","aarshabh",True,1,2}
print("satvik" in set1)
print("king"in set1)
```

```python
#once a set is created you cannot replace its items but you can add new
set1 = {"aanjneya","satvik","aarshabh",True,1,2}
set1.add("king")
print(set1)
```

```python
#once a set is created you cannot replace its items but you can remove
set1 = {"aanjneya","satvik","aarshabh",True,1,2}
set1.remove("aanjneya")
print(set1)
```

```python
#set operations
A = {1,2,3,4,5}
B = {4,5,6,7,8}
print ("union",A|B)
print("intersection",A&B)
print("Difference",A-B)
print("symmetric difference",A^B)
```

```python
#froszen set
x = frozenset({"aanjneya","satvik","aarshabh"})
print(x)
print(type(x))
```

```python
#adding or removing of elements are not allowed in frozen set
x = frozenset({"aanjneya","satvik","aarshabh"})
x.add("king")
x.remove("aanjneya")
print(x)
print(type(x))
```

```python
#changing / adding key:value in dictionary
actor = {"Name": "Shahrukh Khan","YOB": 1965,"Occupation": "Actor"}
actor["super hit"] = "DDLJ"
x = actor.values()
print(x)
```

```python
#to find out a specific value
actor = {"Name": "Shahrukh Khan","YOB": 1965,"Occupation": "Actor"}
print("Name: ",actor["Name"])
```

```python
#adding or upating second element
actor = {"Name": "Shahrukh Khan","YOB": 1965,"Occupation": "Actor"}
actor["Occupation"] = "producer"
actor["YOB"] = 1960
print(actor)
```

```python
#emoving the elements
actor = {"Name": "Shahrukh Khan","YOB": 1965,"Occupation": "Actor"}
actor.pop("YOB")
print("after removal :",actor)
```
